**QUESTION 5**

```python
import pandas as pd
Sales = pd.read_csv("/content/SalesData.csv")
Sales["Sale_amt"]=Sales["Sale_amt"].replace("[/$,]","",regex=True).astype(float)
Sales_data=Sales.groupby("Region")
Sales_data=Sales_data.agg({"SalesMan":"count","Sale_amt":"sum"}).reset_index()
Sales_data.rename(columns={"SalesMan":"salesmen_count","Sale_amt":"total_sales"})
```

**QUESTION 6**

```python
import pandas as pd
Sales = pd.read_csv("/content/SalesData.csv")
Sales["Sale_amt"]=Sales["Sale_amt"].replace("[/$,]","",regex=True).astype(float)
manager_sales = Sales.groupby('Manager')
manager_sales=manager_sales['Sale_amt'].sum().reset_index()
total_sales = manager_sales['Sale_amt'].sum()
manager_sales['percent_sales'] = (manager_sales['Sale_amt'] / total_sales) * 100
manager_sales
```
