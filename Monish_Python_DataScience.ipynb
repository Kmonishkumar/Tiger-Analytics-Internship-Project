**QUESTION 8**

```python
imdb = pd.read_csv("/content/imdb.csv")
imdb["duration"] = pd.to_numeric(imdb["duration"], errors='coerce')
imdb_clean = imdb.dropna(subset=["duration"])
min_duration_index = imdb_clean["duration"].idxmin()
max_duration_index = imdb_clean["duration"].idxmax()
title_of_shortest_movie = imdb_clean.loc[min_duration_index, "title"]
title_of_longest_movie = imdb_clean.loc[max_duration_index, "title"]

print(f"The shortest duraion  movie is: {title_of_shortest_movie}")
print(f"The longest duration movie is: {title_of_longest_movie}")
```

**QUESTION 14**

```python
diamonds_df = pd.read_csv("/content/diamonds.csv")
def compute_volume(row):
    if row['depth'] > 60:
        return row['x'] * row['y'] * row['z']
    else:
        return 8

diamonds_df['volume'] = diamonds_df.apply(compute_volume, axis=1)
diamonds_df
```
